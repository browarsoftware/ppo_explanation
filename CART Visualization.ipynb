```python
import sys
import os
%config InlineBackend.figure_format = 'retina' # Make visualizations look good
#%config InlineBackend.figure_format = 'svg'
%matplotlib inline

if 'google.colab' in sys.modules:
  !pip install -q dtreeviz

import pandas as pd
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

import dtreeviz

random_state = 1234 # get reproducible trees
```

```python
#dataset_url = '/content/96x96_hidden16_power=1_threshold=0.1_1000.txt'
dataset_url = '/content/simple_64x64_hidden16_power=1_threshold=0.3_1000.txtgj.txt'
dataset_url = '/content/64x64_hidden16_power=1_threshold=0.3_1000.txtgj.txt'
dataset = pd.read_csv(dataset_url)
```

```python
feature_names = dataset.columns[1:37].to_list()
target_names = dataset.columns[39]
X = dataset.iloc[:, 1:37].to_numpy()
# jump
y = dataset.iloc[:, 39].to_numpy().astype('int32')
target_names = ['Ground','Jump']

# front back
# y = dataset.iloc[:, 37].to_numpy().astype('int32')
# target_names = ['Stop','Front','Back']

# Left Right
#y = dataset.iloc[:, 38].to_numpy().astype('int32')
#target_names = ['Stop','Left','Right']

"""
tree_classifier = DecisionTreeClassifier(max_depth=4, random_state=random_state,
                                         min_samples_split=30, min_samples_leaf=30)
"""
tree_classifier = DecisionTreeClassifier(random_state=random_state)

tree_classifier.fit(X, y)

viz_model = dtreeviz.model(tree_classifier,
                           X_train=X, y_train=y,
                           feature_names=feature_names,
                           target_name=target_names, class_names=target_names)

import logging
logging.getLogger('matplotlib.font_manager').setLevel(level=logging.CRITICAL)

#viz_model.view(scale=0.8)
#viz_model.view(scale=1.0)
viz_model.view(scale=1.0, label_fontsize=20, ticks_fontsize=9)
#v = viz_model.view(scale=1.0, label_fontsize=20, ticks_fontsize=9)
#v.show()
#v.save("/content/simple64x64hidden16power1threshold0.3.svg")
```
